```python
import matplotlib.pyplot as plt 
import math 
from scipy.stats import norm
import numpy as np 

```

```python
def PDF(n , mean , sd):
    pi = math.pi
    den = sd *( (2 * pi) ** 0.5 )
    po = -1/2*((n- mean)/sd) ** 2
    return math.exp(po) / den
```

```python
mean = 0
sd = 1
limit = 4
x = np.linspace(-limit, limit, 1000)
y = [PDF(i , mean=mean , sd=sd) for i in x]
plt.plot(x , y)
```
